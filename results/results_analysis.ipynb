```python
import pandas as pd
import matplotlib.pyplot as plt

chrome_data = pd.read_csv('test_chrome.csv')
safari_data = pd.read_csv('test_safari.csv')


chrome_data['Time_seconds'] = (chrome_data['Time'] - chrome_data['Time'].iloc[0]) / 1e9
safari_data['Time_seconds'] = (safari_data['Time'] - safari_data['Time'].iloc[0]) / 1e9

plt.figure(figsize=(12, 6))
plt.plot(chrome_data['Time_seconds'], chrome_data['SYSTEM_POWER (Watts)'], label='Chrome PDF Reader', color='blue', linestyle='-', marker='o', markersize=6, linewidth=2)
plt.plot(safari_data['Time_seconds'], safari_data['SYSTEM_POWER (Watts)'], label='Safari PDF Reader', color='red', linestyle='--', marker='x', markersize=6, linewidth=2)
plt.title('System Power Consumption (Watts) Over Time')
plt.xlabel('Time (seconds)')
plt.ylabel('Power (Watts)')
plt.legend()
plt.grid(True)
plt.show()

chrome_data['avg_cpu_usage'] = chrome_data[[f'CPU_USAGE_{i}' for i in range(10)]].mean(axis=1)
safari_data['avg_cpu_usage'] = safari_data[[f'CPU_USAGE_{i}' for i in range(10)]].mean(axis=1)

plt.figure(figsize=(12, 6))
plt.plot(chrome_data['Time_seconds'], chrome_data['avg_cpu_usage'], label='Chrome PDF Reader', color='blue', linestyle='-', marker='o', markersize=6, linewidth=2)
plt.plot(safari_data['Time_seconds'], safari_data['avg_cpu_usage'], label='Safari PDF Reader', color='red', linestyle='--', marker='x', markersize=6, linewidth=2)
plt.title('Average CPU Usage Over Time')
plt.xlabel('Time (seconds)')
plt.ylabel('CPU Usage (%)')
plt.legend()
plt.grid(True)
plt.show()
```
